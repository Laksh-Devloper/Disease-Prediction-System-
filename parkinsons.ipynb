```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
data = pd.read_csv("M:\\Techsaksham_by_Microsoft_Project\\Project\\dataset\\parkinsons_tomeklinks.csv")
```

```python
X = data.drop(columns=['status'] , axis=1)
Y = data['status']
```

```python
print(X , '-L-' , Y)
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size=0.2 , random_state=2)
```

```python
print(X.shape , Y.shape)
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
rf_model.fit(X_train, Y_train)
```

```python
y_pred_rf = rf_model.predict(X_test)
tr_accuracy = accuracy_score(Y_test, y_pred_rf)
print(tr_accuracy)
```

```python
# Confusion Matrix
cm = confusion_matrix(Y_test, y_pred_rf)
print("Confusion Matrix :- \n" , cm)

# Precision
precision = precision_score(Y_test, y_pred_rf)  # TP / (TP + FP)
print("Precision:", precision)

# Recall
recall = recall_score(Y_test, y_pred_rf)  # TP / (TP + FN)
print("Recall:", recall)

# F1-Score (Harmonic mean of precision & recall)
f1 = f1_score(Y_test, y_pred_rf)
print("F1-Score:", f1)
```
